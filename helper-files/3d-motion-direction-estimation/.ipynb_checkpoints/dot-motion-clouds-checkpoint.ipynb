```python
import numpy as np
import matplotlib.pyplot as plt
from dotMotion import CoherentMotionFactory, TrueProjector
```

```python
nDots = 15
cubeSize = 10 # cm
motionGenerator = CoherentMotionFactory(cubeSize,nDots)

speed = .1 # pixels per frame
rng = np.random
theta = 0
phi = 90
totFrames = 100
X,Y,Z = motionGenerator.generate(rng, speed, theta, phi, totFrames)
```

```python
# Define projection parameters
screenDist = 100.0   # Distance from the screen
ipd = 6.5            # Inter-pupillary distance
eye = (5, 5)         # Eye position (x, y)

# Compute the 2D projections for the left and right eye
lx_p, rx_p, y_p = TrueProjector.project(X, Y, Z, screenDist, ipd, eye)
```

```python
plt.plot(lx_p.T)
plt.show()
```
